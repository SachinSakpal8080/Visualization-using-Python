```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(color_codes=True)

import warnings
warnings.filterwarnings('ignore')
```

```python
#import data
data = pd.read_csv('Automobile_visualization_using_python.csv')
```

```python
data.head().T
```

```python
sns.distplot(data['normalized_losses'], kde=False)
plt.show()
```

```python
sns.distplot(data['normalized_losses'], kde=True)
plt.show()
```

```python
# sns.distplot(data['fuel_type'], kde=False)
# plt.show()
# We need to give the int or float type feature.
```

```python
sns.distplot(data['city_mpg'], kde=True)
plt.show()
# With KDE line
```

```python
sns.distplot(data['city_mpg'], kde=False, rug=True)
plt.show()
# Without KDE line
```

```python
sns.jointplot(data['engine_size'], data['horsepower'])
plt.show()
```

```python
sns.jointplot(data['engine_size'], data['horsepower'], kind='hex')
plt.show()
```

```python
sns.jointplot(data['engine_size'], data['horsepower'], kind='residual')
plt.show()
```

```python
sns.jointplot(data['engine_size'], data['horsepower'], kind='reg')
plt.show()
```

```python
sns.jointplot(data['engine_size'], data['horsepower'], kind='kde')
plt.show()
```

```python
sns.pairplot(data[['normalized_losses', 'engine_size', 'horsepower']])
plt.show()
```

```python
sns.stripplot(data['fuel_type'], data['horsepower'], jitter=True)
plt.show()
```

```python
sns.swarmplot(data['fuel_type'], data['horsepower'])
plt.show()
```

```python
sns.boxplot(data['fuel_type'], data['horsepower'])
plt.show()
```

```python
sns.boxplot(data['number_of_doors'], data['horsepower'])
plt.show()
```

```python
sns.boxplot(data['number_of_doors'], data['horsepower'], hue=data['fuel_type'])
plt.show()
```

```python
sns.barplot(data['body_style'], data['horsepower'])
plt.show()
```

```python
sns.barplot(data['body_style'], data['horsepower'], hue=data['engine_location'])
plt.show()
```

```python
sns.countplot(data['body_style'])
plt.show()
```

```python
sns.countplot(data['body_style'], hue=data['engine_location'])
plt.show()
```

```python
sns.pointplot(data['fuel_system'], data['horsepower'])
plt.show()
```

```python
sns.pointplot(data['fuel_system'], data['horsepower'], hue=data['number_of_doors'])
plt.show()
```

```python
sns.factorplot(x = 'fuel_type', y = 'horsepower', 
               hue = 'number_of_doors', col = 'engine_location', 
               data = data, kind = 'swarm')
plt.show()
```

```python
sns.factorplot(x = 'fuel_type', y = 'horsepower', 
               hue = 'number_of_doors', col = 'engine_location', 
               data = data, kind = 'bar')
plt.show()
```

```python
sns.factorplot(x = 'fuel_type', 
               hue = 'number_of_doors', col = 'engine_location', 
               data = data, kind = 'count')
plt.show()
```

```python
sns.factorplot(x = 'fuel_type', y = 'horsepower', 
               hue = 'number_of_doors', col = 'engine_location', 
               data = data, kind = 'box')
plt.show()
```

```python
sns.factorplot(x = 'fuel_type', y = 'horsepower', 
               hue = 'number_of_doors', col = 'engine_location', 
               data = data, kind = 'violin')
plt.show()
```
